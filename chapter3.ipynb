3.1 Seconds in an Hour
There are 3,600 seconds in an hour. This is calculated by multiplying the number of seconds in a minute (60) by the number of minutes in an hour (60).

```python
60 * 60
```

3.2 Seconds per Hour Variable
Assigning the result of seconds in an hour to a variable called seconds_per_hour.

```python
seconds_per_hour = 60 * 60
seconds_per_hour
```

3.3 Seconds in a Day
To calculate the number of seconds in a day, we multiply seconds_per_hour by the number of hours in a day (24).

```python
seconds_per_hour * 24
```

3.4 Seconds per Day Variable
Assigning the result of seconds in a day to a variable called seconds_per_day.

```python
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

3.5 Seconds per Day divided by Seconds per Hour
Dividing seconds_per_day by seconds_per_hour using floating-point division. This should return 24.0, the number of hours in a day.

```python
seconds_per_day / seconds_per_hour
```

3.6 Integer Division of Seconds per Day by Seconds per Hour
Dividing seconds_per_day by seconds_per_hour using integer division. Yes, the result agrees with the floating-point value from 3.5, the only difference is that integer division returns 24 instead of 24.0, dropping the decimal point.

```python
seconds_per_day // seconds_per_hour
```
